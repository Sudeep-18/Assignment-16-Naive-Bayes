```python
df = df.drop_duplicates()
```

```python
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df['native'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
cols = ['age','educationno','capitalgain','capitalloss']
scaler = MinMaxScaler()
df[cols]=scaler.fit_transform(df[cols])
df.head()
```

```python
# Label encoding 
cols = ['workclass','education','maritalstatus','occupation','relationship','race','sex','hoursperweek','native','Salary']
encoder = LabelEncoder()
for i in cols:
    df[i] = encoder.fit_transform(df[i])
df.head()
```

```python
df.Salary.unique()
```

## Naive Bayes

```python
from sklearn.model_selection import train_test_split
x = df.iloc[:,0:-1]
y= df['Salary']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.naive_bayes import CategoricalNB as CB
import numpy as np
```

```python
classifier_cb = CB()
classifier_cb.fit(x_train,y_train)
```

```python
train_pred_c = classifier_cb.predict(x_train)
```

```python
train_acc = np.mean(train_pred_c==y_train)
train_acc
```

```python
test_pred_c= classifier_cb.predict(x_test)
```

```python
test_pred_c
```

```python
test_acc = np.mean(test_pred_c==y_test)
```

```python
test_acc
```
